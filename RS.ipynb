# A Graph-based system

```python
import networkx as nx
from networkx import info
```

```python
anime = anime.loc[anime['aired_from_year']>=2000] # we only focus on animes after 2016
```

```python
anime['aired_from_year'].value_counts()
```

```python
users = users.sample(n=500) # randomly select 500 users
```

```python
ual = ual.loc[ual['username'].isin(users.username.unique())]
```

```python
ual.info()
```

```python
B = nx.DiGraph()

# Add nodes
B.add_nodes_from(anime['anime_id'].unique(),node_type='anime')         
B.add_nodes_from(users['username'].unique(),node_type='user')
```

```python
info(B)
```

```python
all_genres = set()
for i, row in anime[['anime_id','genre']].iterrows():
    if pd.notnull(row['genre']):
        genres = row['genre'].split(',')
        for genre in genres:
            genre = genre.strip()
            if genre not in all_genres:
                all_genres.add(genre)
                B.add_node(genre,node_type='genre')
            B.add_edge(row['anime_id'],genre)
```

```python
info(B)
```

```python
# Each anime producer specializes a few types of and attrats certain group of audience, user bondiness to the same producer/studio

all_studios = set()
for i, row in anime[['anime_id','studio']].iterrows():
    if pd.notnull(row['studio']):
        studios = row['studio'].split(',')
        for studio in studios:
            studio = studio.strip()
            if studio not in all_studios:
                all_studios.add(studio)
                B.add_node(studio,node_type='studio')
            B.add_edge(row['anime_id'],studio)
```

```python
all_producers = set()
for i, row in anime[['anime_id','producer']].iterrows():
    if pd.notnull(row['producer']):
        producers = row['producer'].split(',')
        for producer in producers:
            producer = producer.strip()
            if producer not in all_producers:
                all_producers.add(producer)
                B.add_node(producer,node_type='producer')
            B.add_edge(row['anime_id'],producer)
```

```python
# Season, assume people would be interested in animes that were in the same period as they liked/rated highly
# reduce the bias brought by popularity

lookup = {'Winter':'1','Spring':'2','Summer':'3','Fall':'4'}

season_lookup = {season:'-'.join([season.split()[1], lookup[season.split()[0]]]) for season in anime['premiered'].unique() if pd.notnull(season)}

sorted_season_list = sorted(list(season_lookup.values())) # chronological sequence of seasons
for i, season_code in enumerate(sorted_season_list):
    B.add_node(season_code, node_type='season')
    if i>=1:
        B.add_edge(season_code, sorted_season_list[i-1]) # connect consequent seasons together

for anime_id, season in anime[['anime_id','premiered']].values:
    if pd.notnull(season):
        B.add_edge(anime_id,season_lookup[season]) # edges that connect anime to its season
```

```python
ual = ual.dropna(subset=['username','anime_id','my_score'])
```

```python
ual['score_normalized'] = ual.groupby('username')['my_score'].transform(lambda x: (x - x.mean()) / x.std())
```

```python
for user, anime_id, score, score_norm in ual[['username','anime_id','my_score','score_normalized']].values:
    if score>=5 and anime_id in list(B.nodes):
        B.add_edge(user, anime_id, weight=score_norm+1)
```

```python
B.remove_nodes_from(list(nx.isolates(B)))
```

# Script code

```python
class User:
    def __init__(self, username):
        self.username = username
        self.viewed_animes = []
    
    def get_viewed_anime_list(self):
        return self.viewed_animes
```

```python
class Recommender:
    def __init__(self):
        self.graph = None
        pass

    def build_graph(self, users, animes, ual):
        B = nx.DiGraph()
        
        # Add nodes
        B.add_nodes_from(animes['anime_id'].unique(),node_type='anime')         
        B.add_nodes_from(users['username'].unique(),node_type='user')
        
        all_genres = set()
        for i, row in animes[['anime_id','genre']].iterrows():
            if pd.notnull(row['genre']):
                genres = row['genre'].split(',')
                for genre in genres:
                    genre = genre.strip()
                    if genre not in all_genres:
                        all_genres.add(genre)
                        B.add_node(genre,node_type='genre')
                    B.add_edge(row['anime_id'],genre)
        
        all_studios = set()
        for i, row in anime[['anime_id','studio']].iterrows():
            if pd.notnull(row['studio']):
                studios = row['studio'].split(',')
                for studio in studios:
                    studio = studio.strip()
                    if studio not in all_studios:
                        all_studios.add(studio)
                        B.add_node(studio,node_type='studio')
                    B.add_edge(row['anime_id'],studio)
    
    def run_user(self, active_user, damping_factor):
        ppr_values = nx.pagerank(G=self.graph, alpha=damping_factor, max_iter=1000, personalization={active_user:1.0}, weight='weight')
        anime_ppr_values = {node:value for node,value in ppr_values.items() if nx.get_node_attributes(B, "node_type")[node]=='anime' 
                            and node not in active_user.get_viewed_anime_list()}
        return topn_from_dict(anime_ppr_values, 10)
```
